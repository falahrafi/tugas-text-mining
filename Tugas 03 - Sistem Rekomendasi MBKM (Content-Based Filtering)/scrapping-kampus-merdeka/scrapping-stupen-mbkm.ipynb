```python
import requests
```

<br>
<h3><b>Ambil ID dari Sector (Bidang) yang Diinginkan</b></h3>

```python
url = "https://api.kampusmerdeka.kemdikbud.go.id/mitra/sector"
r = requests.get(url)
sectors = r.json()
```

```python
for sector in sectors['data']:
  print(sector)
  
  if(sector['name'] == 'Technology'):
    sector_tech_id = sector['id']
```

<br>
<h3><b>Ambil ID dari Course & Mitra yang Termasuk Bidang Teknologi</b></h3>

```python
url = "https://api.kampusmerdeka.kemdikbud.go.id/studi/browse/activity?sector_id=" + sector_tech_id
r = requests.get(url)
courses = r.json()
```

```python
courses['data'][0]
```

```python
courses_dict = {'mitra_id' : [],
                'course_id': []}

for course in courses['data']:
  courses_dict['mitra_id'].append(course['mitra_id'])  
  courses_dict['course_id'].append(course['id'])
```

```python
import pandas as pd
df_course = pd.DataFrame(courses_dict)
```

```python
df_course
```

<br>
<h3><b>Ambil Nama & Alamat Mitra</b></h3>

```python
mitra_details = []

for mitra_id in df_course['mitra_id']:
  url = "https://api.kampusmerdeka.kemdikbud.go.id/mitra/public/id/" + mitra_id
  r = requests.get(url)
  mitra_detail = r.json()
  mitra_details.append(mitra_detail)
```

```python
mitra_details[9]
```

```python
mitra_dict = {'mitra_name': [],
             'address': []}

for md in mitra_details:    
  mitra_dict['mitra_name'].append(md['data']['name'])
  mitra_dict['address'].append(md['data']['hq_address'])
```

<br>
<h3><b>Ambil Data Modul Pembelajaran dari Tiap Course</b></h3>

```python
course_details = []

for course_id in df_course['course_id']:
  url = "https://api.kampusmerdeka.kemdikbud.go.id/studi/browse/activity/" + course_id
  r = requests.get(url)
  course_detail = r.json()
  course_details.append(course_detail)
```

```python
course_details[9]
```

```python
desc_dict = {'description': [],
            'course_name': []}

for cd in course_details:
  
  desc_text = ""

  for module in cd['data']['module']:
    desc_text += module['objective'] + ' '
    
  desc_dict['description'].append(desc_text)
  desc_dict['course_name'].append(cd['data']['name'])  
```

<br>
<h3><b>Export Dataset (Bahasa Indonesia)</b></h3>

```python
import pandas as pd

df_stupen = df_course.copy()
df_stupen["mitra_name"] = mitra_dict['mitra_name']
df_stupen["course_name"] = desc_dict['course_name']
df_stupen["address"] = mitra_dict['address']
df_stupen["description"] = desc_dict['description']

del df_stupen["mitra_id"]
del df_stupen["course_id"]

df_stupen.to_csv('CSV/dataset_stupen.csv', index=False)

df_stupen
```

<br>
<h3><b>Export Dataset (Bahasa Inggris)</b></h3>

```python
# pip install -U deep-translator
```

```python
import nltk
from deep_translator import GoogleTranslator

df_stupen_eng = df_stupen.copy()

for i, desc in enumerate(df_stupen_eng['description']):

  desc_translated = ""
  desc_token = nltk.tokenize.sent_tokenize(desc)

  # Mengatasi batasan jumlah karakter deep translator
  for sentence in desc_token:
    desc_translated += GoogleTranslator(source='auto', target='en').translate(sentence)

  df_stupen_eng.at[i, 'description'] = desc_translated 
```

```python
df_stupen_eng.to_csv('CSV/dataset_stupen_english.csv', index=False)
df_stupen_eng
```
