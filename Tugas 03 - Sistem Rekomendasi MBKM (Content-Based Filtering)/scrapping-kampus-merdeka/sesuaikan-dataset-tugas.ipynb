```python
import pandas as pd

filenames = [
              "CSV/dataset_falah_english.csv",
              "CSV/dataset_stupen_english.csv",
              "CSV/dataset_magang_english.csv"
            ]

# merging two csv files
df = pd.concat(
    map(pd.read_csv, filenames), ignore_index=True)

df
```

<br>
<h3><b>Menentukan Jenis Program + Menggabungkan Nama Kursus & Posisi</b></h3>

```python
import math

program_dict = {'program': [],
               'course_position': []}

for c, p in zip( df['course_name'], df['position_name'] ):

  if isinstance(c, str):
    program = "Studi Independen"
    program_dict['course_position'].append(c)
  elif isinstance(p, str):
    program = "Magang"
    program_dict['course_position'].append(p)
  else:
    program = "Kuliah"
    program_dict['course_position'].append("Mata Kuliah")
  
  program_dict['program'].append(program)
```

<br>
<h3><b>Memberi Keterangan pada Alamat yang Tidak Tersedia</b></h3>

```python
new_address_dict = {'address': []}

for a in df['address']:

  if isinstance(a, float) or a ==  "-":
    new_address_dict['address'].append("(Alamat Tidak Tersedia)")
  else:
    new_address_dict['address'].append(a)
```

<br>
<h3><b>Merapikan Alamat</b></h3>

```python
one_line_address_dict = {'address': []}

for a in new_address_dict['address']:
  a = " ".join(a.split("\n"))
  one_line_address_dict['address'].append(a)
```

<br>
<h3><b>Export CSV</b></h3>

```python
df_new = df.copy()

df_new['program'] = program_dict['program']
df_new['course_position'] = program_dict['course_position']

del df_new['course_name']
del df_new['position_name']

del df_new['address']
df_new['address'] = one_line_address_dict['address']

# Ubah urutan kolom
df_new = df_new[['mitra_name', 'address','description', 'program', 'course_position']]
```

```python
df_new.to_csv('CSV/dataset_tugas_falah_english.csv', index=False)
df_new
```
